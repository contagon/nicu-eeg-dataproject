```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mne
from ipywidgets import *
%matplotlib notebook
```

```python
raw = mne.io.read_raw_edf('data/eeg1.edf')
```

```python
#get data
time = raw[0,:][1]
signal = np.squeeze(raw[0,:][0])
seizure = np.genfromtxt(open("data/annotations_2017_A.csv", 'rb'), delimiter=',',skip_header=1)
seizure = seizure[:6993,1]

upper = np.argwhere(seizure[:-1] - seizure[1:] == 1)
lower = np.argwhere(seizure[:-1] - seizure[1:] == -1)

fig = plt.figure(figsize=(25,5))
ax = fig.add_subplot(1, 1, 1)
for i,j in zip(lower, upper):
    ax.axvspan(i,j,alpha=0.2,color='r')
line, = ax.plot(time, signal)

@interact
def update(w=0.1):
    ax.set_xlim(w-50, w+50)
    fig.canvas.draw()
    fig.canvas.flush_events()
```

```python
t = np.linspace(-5,5,100)
ax = plt.subplot(111)
ax.plot(t, (t)**2)

def update(a):
    ax.set_xlim(a-10, a+10)
    
interact(update, a=widgets.FloatSlider(min=0, max=1, step=.001, value=50))
```
